```python
def draw_boxes(image, boxes, class_names, scores, fname, max_boxes=10, min_score=0.1):
  """Overlay labeled boxes on an image with formatted scores and label names."""
  colors = list(ImageColor.colormap.values())

  df_All = pd.DataFrame(columns = ['xMin', 'yMin', 'width', 'height', 'Total_Faces', 'Image_Name'])

  try:
    font = ImageFont.truetype("/usr/share/fonts/truetype/liberation/LiberationSansNarrow-Regular.ttf",
                              25)
  except IOError:
    print("Font not found, using default font.")
    font = ImageFont.load_default()

  count = 0
  for i in range(min(boxes.shape[0], max_boxes)):
    if scores[i] >= min_score and class_names[i].decode("ascii") == 'Human face':
      count = count+1


  for i in range(min(boxes.shape[0], max_boxes)):
    if scores[i] >= min_score and class_names[i].decode("ascii") == 'Human face':
      ymin, xmin, ymax, xmax = tuple(boxes[i])
      df2 = {'xMin': xmin, 'yMin': ymin, 'width': xmax-xmin, 'height':ymax-ymin, 'Total_Faces':int(count), 'Image_Name':fname}
      df_All = df_All.append(df2, ignore_index = True)

  df_All.to_csv('output.csv', mode='a', header=False)
  return image
```
